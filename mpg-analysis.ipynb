```python
import seaborn as sns
```

```python
mpg = sns.load_dataset("mpg")
mpg
```

```python
###Create a visual showing the relationship between horsepower and fuel economy.
```

```python
sns.relplot(data=mpg, x = "horsepower", y = "mpg", hue="model_year") 
```

```python
#As the newer models are made the mpg get efficiently better over time compare to the older models.
```

```python
###Create a table summarizing fuel economy by country of origin
```

```python
fuelbyorigin=mpg.groupby("name").agg({"mpg":"sum","origin":"sum"})

fuelbyorigin.head(10)
```

```python
###Create a visual showing how fuel economy has changed over time
```

```python
sns.relplot(data=mpg, x = "model_year", y = "mpg",kind="line")
```

```python
###Create two other visuals or tables related to fuel economy
```

```python
sns.displot(data=mpg, x = "cylinders", y = "mpg", hue="model_year") 
```

```python
sns.displot(data=mpg, x = "acceleration", y = "mpg")
```
